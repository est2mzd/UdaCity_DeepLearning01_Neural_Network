```python
# Solution is available in the other "solution.py" tab
import tensorflow as tf

output = None

# Input : 3x4
inputs = [[1.0, 2.0, 3.0, 4.0], 
          [-1.0, -2.0, -3.0, -4.0], 
          [11.0, 12.0, 13.0, 14.0]]

# weight_1 : 4x3
hidden_layer_weights = [
    [0.1, 0.2, 0.4],
    [0.4, 0.6, 0.6],
    [0.5, 0.9, 0.1],
    [0.8, 0.2, 0.8]]

# weight_2 : 3x2
out_weights = [
    [0.1, 0.6],
    [0.2, 0.1],
    [0.7, 0.9]]

# Weights and biases
features = tf.placeholder(tf.float32, [3, 4]);
weight_1 = tf.Variable(tf.truncated_normal( [4, 3] ) );
weight_2 = tf.Variable(tf.truncated_normal( [3, 2] ) );
#
bias_1 = tf.Variable(tf.zeros(3));
bias_2 = tf.Variable(tf.zeros(2));


# TODO: Create Model
hidden_layer = tf.add( tf.matmul( features, weight_1), bias_1 )
hidden_layer = tf.nn.relu( hidden_layer )
logits       = tf.add( tf.matmul( hidden_layer, weight_2 ), bias_2 )
#
init = tf.global_variables_initializer()
```

```python
# TODO: Print session results
with tf.Session() as sess:
    sess.run(init)
    output = sess.run(logits, feed_dict = {features : inputs})
    print(output)
    
```
